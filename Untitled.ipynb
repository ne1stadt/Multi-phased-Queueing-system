```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from mpl_toolkits import mplot3d
import time
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
def model_theta(initial_lambda, mus, n, tag = 'nediff', detalization = 20):
    #initial checks
    if (0 in mus):
        print('Mu must be greater then 0 in all nodes')
    
    right_border = 10/initial_lambda
    x = np.linspace(0, right_border, detalization)
    data = pd.DataFrame({'grid': x})
    
    if (tag == 'diff'):
        data['generator'] = initial_lambda*np.exp(-1*initial_lambda*x)
    else:
        data['generator'] = distribution = 1-np.exp(-1*initial_lambda*x)
    
    lambd = initial_lambda
    rho = 0
    phi = 0
    coefs = pd.DataFrame({'lambda': [lambd], 'mu': [0], 'rho': [rho], 'phi': [phi], 'cosh2phi': [1], 'sinh2phi': [0]})
    node_number = 1
    for mu in mus:
        lambd = lambd*((math.cosh(phi))**2) - mu*((math.sinh(phi))**2)
        rho = lambd/mu
        phi = math.atanh(rho**((n+1)/2))
        coefs = pd.concat([coefs, pd.DataFrame([[lambd, mu, rho, phi, (math.cosh(phi))**2, (math.sinh(phi))**2]], columns = ['lambda', 'mu', 'rho', 'phi', 'cosh2phi', 'sinh2phi'])], ignore_index = True)
        if (tag == 'diff'):
            data['node_' + str(node_number)] = lambd*((math.cosh(phi))**2)*np.exp(-1*lambd*x) - mu*((math.sinh(phi))**2)*np.exp(-1*mu*x)
        else:
            data['node_' + str(node_number)] = 1-((math.cosh(phi))**2)*np.exp(-1*lambd*x) + ((math.sinh(phi))**2)*np.exp(-1*mu*x)
        node_number = node_number + 1
        
    return data, coefs
```

```python
def interprete(data, lam, mus, queue_length, detalization, i): #takes data from simulation and returns dataframe with outcome distributions on node i
    model_data, model_coefs = model_theta(lam, mus, queue_length, 'diff', detalization)
    plt.hist(distr_substract(data, i), bins=detalization, density = True, label = 'simulation on node ' + str(i))
    if (i == 0):
        plt.plot(model_data['grid'], model_data['generator'], label = 'generator')
    else:
        plt.plot(model_data['grid'], model_data['node_' + str(i)], label = 'theta model on node ' + str(i))
    plt.title('Result on node ' + str(i))
    print('Model')
    print('Arrival rate: ' + str(model_coefs.loc[i, 'lambda']))
    print('Mean Service Time: ' + str(model_coefs.loc[i, 'mu']))
    print('Coefficient rho: ' + str(model_coefs.loc[i, 'rho']))
    print('Coefficient phi: ' + str(model_coefs.loc[i, 'phi']))
    print('ch2phi = ' + str(model_coefs.loc[i, 'cosh2phi']) + ' and ' + 'sh2phi = ' + str(model_coefs.loc[i, 'sinh2phi']))
    print()
    print('Simulation results')
    print()
    plt.show()
```

```python
def distr_substract(data, k): #substracts previous value for each element in array for node k
    if (k == 0):
        array = data.drop([], axis = 0)['start_time'].to_numpy()
    else:
        drop_indexes = data[data['innode'] < k].index
        array = data['node_' + str(k) + '_end'].drop(drop_indexes, axis = 0).to_numpy()
    
    l = len(array)
    for i in range(l-1, 0, -1):
        array[i] = array[i] - array[i-1]
    return array
```

```python
def build(data, lam, mus, queue_length, detalization):
    m = len(mus)
    for i in range(m+1):
        interprete(data, lam, mus, queue_length, detalization, i)
```
